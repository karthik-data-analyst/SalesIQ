```python
import pandas as pd

# ---- LOAD DATA ----
df = pd.read_csv(r"D:\Data Analytics\Data Analyst Project\SalesIQ\Dataset\Superstore.csv", 
                 encoding='latin1')

print("Raw Data Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 3 rows:")
print(df.head(3))
```

```python

print("Raw Data Shape:", df.shape)

# ---- CLEAN COLUMN NAMES ----
# Remove spaces from column names
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ---- FIX DATES ----
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='mixed')

# ---- CREATE NEW COLUMNS ----
# These are what AI and Power BI need
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month
df['Order_Month_Name'] = df['Order_Date'].dt.strftime('%B')
df['Profit_Margin'] = (df['Profit'] / df['Sales'] * 100).round(2)

# ---- CHECK FOR NULLS ----
print("\nNull values per column:")
print(df.isnull().sum())

# ---- REMOVE DUPLICATES ----
before = len(df)
df = df.drop_duplicates()
after = len(df)
print(f"\nDuplicates removed: {before - after}")



print("\nClean Data Shape:", df.shape)
print("\nNew Columns Added:")
print(['Order_Year', 'Order_Month', 'Order_Month_Name', 'Profit_Margin'])
print("\n✅ Data cleaning complete! superstore_clean.csv saved.")
```
